# Matplotlib plotting tools

<mark>Matplotlib</mark> is the standard Python library to create plots. It works best with <mark>numpy</mark>, a numerical library in Python. Using these libraries together, one can create many plots.

We'll learn more about Numpy in future sessions.

```python
# You can import matplotlib and numpy as follows
import numpy as np
import matplotlib.pyplot as plt

# If you're working on a jupyter notebook include the following line in the notebook:
%matplotlib inline
```

```python
f = lambda x : x**3 + 2*x -6

x=np.linspace(0,10,endpoint=True)
```

```python
plt.figure()
plt.plot(x,f(x))
plt.show()
```

### Graphing Functions

```python
# First, create a figure and subplot from 1x1 grid
# You can change the grid by changing the first two numbers in the subplot command
plt.figure(figsize=(8,6))
plt.subplot(1,1,1)

# Create an array with the arguments for the domain of the function
x=np.linspace(-2*np.pi,2*np.pi,1000,endpoint=True)

# Create the array with function values
# Notice the argument is a numpy array
f=np.cos(x)

# Adjust the limits so that everything is more comfortable
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(f.min() * 1.1, f.max() * 1.1)

# Create the plot
plt.plot(x,f)

# Move axes to traditional place in math
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# Make titles and labels
plt.title('$y=\mathrm{cos}(x)$')
#plt.xlabel('x-axis') #uncomment if you want axis labels, usually works better with default axes
#plt.ylabel('y-axis') #uncomment if you want axis labels, usually works better with default axes

# Changing ticks to show important values
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],
           [r'$-2\pi$',r'$-\pi$',r'$0$',r'$\pi$',r'$2\pi$'])
plt.yticks([-1,0,1],
          [r'$-1$',r'$0$',r'$1$'])

# Locate and decorate a special value
m=7*np.pi/4

plt.plot([m,m],[0,np.cos(m)],color='blue',linewidth=1.5,linestyle="--") # creates a line

plt.scatter([m,m],[0,np.cos(m)],15,color='blue') # creates a dot

plt.annotate(r'$\cos(\frac{7\pi}{4})$',
             xy=(m,np.cos(m)) , xycoords='data',
             xytext=(-90,20), textcoords='offset points', fontsize=12,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")) # creates a label and arrow

plt.annotate(r'$x=\frac{7\pi}{4}$',
             xy=(m,0) , xycoords='data',
             xytext=(-30,-40), textcoords='offset points', fontsize=12,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")) # creates a label and arrow

# Ask it to show
plt.show()

# Close the plotting to move on
plt.close()
```

### Graphing More Than One Function

This is relatively simple, you just have to add another array of function values and another call to plot.

```python
plt.figure(figsize=(8,6))
plt.subplot(1,1,1)

x=np.linspace(-2*np.pi,2*np.pi,1000,endpoint=True)

# Make two function value arrays this time
f=np.pi*np.cos(x)
g=np.pi*np.cos(x+np.pi/4)


plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(f.min() * 1.1, f.max() * 1.1)

# Create two plots this time
plt.plot(x,f)
plt.plot(x,g)

plt.title('Cosine Shifts')

# A legend is very useful this time around
plt.plot(x, f, color="blue", linewidth=1.5, linestyle="-", label="$y=\cos(x)$")
plt.plot(x, g, color="green",  linewidth=1.5, linestyle="-", label="$y=\cos(x+\pi/4)$")
plt.legend(loc='upper left')

# Other stuff we've done
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],
           [r'$-2\pi$',r'$-\pi$',r'$0$',r'$\pi$',r'$2\pi$'])
plt.yticks([-np.pi,0,np.pi],
          [r'$-\pi$',r'$0$',r'$\pi$'])

# this gives the labels a slight white background:
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))

    
plt.show()
plt.close()
```

### Using subplots

You can plot multiple graphs on subplots.

```python
# Start with some data to plot
# Here we'll do six closely related functions
x=np.linspace(-3,3,1000,endpoint=True)
f1=np.exp(-x**2)
f2=np.exp(-5*(x**2))
f3=np.exp(-10*(x**2))
f4=np.exp(-20*(x**2))
f5=np.exp(-40*(x**2))
f6=np.exp(-80*(x**2))

# One can create a figure called F and name each of the subplots in it
F, ((fig1, fig2, fig3),(fig4, fig5, fig6)) = plt.subplots(2,3,sharex='col',sharey='row',figsize=(10,8))

# To give the whole collection a title do the following
plt.suptitle('Pulse $y=e^{-kx^2}$', fontsize=16)

# Can assign each of the figures some data
fig1.plot(x,f1)
fig1.set_xlim([-3.1,3.1])
fig1.set_ylim([0, 1.1])
fig1.set_title('$k=1$')

fig2.plot(x,f2)
fig2.set_xlim([-3.1,3.1])
fig2.set_ylim([0, 1.1])
fig2.set_title('$k=5$')

fig3.plot(x,f3)
fig3.set_xlim([-3.1,3.1])
fig3.set_ylim([0, 1.1])
fig3.set_title('$k=10$')

fig4.plot(x,f4)
fig4.set_xlim([-3.1,3.1])
fig4.set_ylim([0, 1.1])
fig4.set_title('$k=20$')

fig5.plot(x,f5)
fig5.set_xlim([-3.1,3.1])
fig5.set_ylim([0, 1.1])
fig5.set_title('$k=40$')

fig6.plot(x,f6)
fig6.set_xlim([-3.1,3.1])
fig6.set_ylim([0, 1.1])
fig6.set_title('$k=80$')
```

### Plotting Parametric Curves

You can plot parametric curves in the plane. That is, curves like the following:
$$
\gamma:[0,25]\to \mathbb{R}^2
\qquad
\gamma(t)=(t\cos(t),t\sin(t))
$$

In fact, if you look closely, functions were plotted before by using the parametrization:
$$
\gamma(t)=\left(t,f(t)\right)
$$

```python
# Data for a spiral
t=np.linspace(0,25,1000,endpoint=True)
x=t*np.cos(t)
y=t*np.sin(t)

# Create a figure
plt.figure()
plt.plot(x,y)

# Moving the axes as before
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.show()
```

You can do scatter plots instead to highlight the points that are being plotted: 

```python
# Data for a spiral
t=np.linspace(0,25,150,endpoint=True)
x=t*np.cos(t)
y=t*np.sin(t)

# Create a figure
F, (ax1,ax2) = plt.subplots(1,2,sharex='col',sharey='row',figsize=(12,6))
ax1.plot(x,y,'ro')
ax2.plot(x,y,'k-')
ax2.plot(x,y,'ro')

plt.show()
```

### Plotting loci of equations or implicit plots

At times you may need to plot the solution set to an equation like the following:
$$
y^2=x^3-2x+1
$$
That is, you want to plot the following set:
$$
\big\{(x,y)\in\mathbb{R}^2\mid y^2=x^3-2x+1\big\}
$$
The trick is to move everything to one side to define a function $f(x,y)=-y^2+x^3-2x+1$, and then use the <mark>contour</mark> function to obtain the $0$-level set of this function.

```python
x = np.linspace(-5., 5.)
y = np.linspace(-5., 5.)[:, None]
plt.contour(x, y.ravel(),-y**2+x**3-2*x+1,[0])
plt.title('The elliptic curve $y^2=x^3-2x+1$')
```

### Two variables functions: contour plots

This leads us to contour plots of two variable functions. That is, given a function like:
$$
f(x,y)=2x^2+3y^2-4
$$
you want to plot the level sets:
$$
\Sigma_c=\big\{(x,y)\in\mathbb{R}^2\mid f(x,y)=c\big\}
$$
for different constants $c$.

```python
x = np.linspace(-3.,3., 100)
y = np.linspace(-3., 3., 100)
X, Y = np.meshgrid(x,y)
Z = 2*X**2+3*Y**2-4

plt.contour(X, Y, Z)
plt.title('Level sets of $f(x,y)=2x^2+3y^2-4$')
plt.show()
```

```python
# You can pick the level sets as follow 
# Note this is what we wanted to do when we wanted to do an implicit plot

plt.contour(X, Y, Z,[-5,-4,-3,-2,-1,6,7,8,9,10])
plt.title('Level sets of $f(x,y)=2x^2+3y^2-4$')
plt.show()
```

There are other styles of contour plots

```python
plt.contourf(X, Y, Z)
plt.title('Contour plot $f(x,y)=2x^2+3y^2-4$')
plt.show()
```

```python
plt.pcolormesh(X, Y, Z)
plt.title('Level sets of $f(x,y)=2x^2+3y^2-4$')
plt.show()
```

And other color maps, as the following shows (look <a href=http://matplotlib.org/examples/color/colormaps_reference.html>here</a> for a list of available color maps).

```python
plt.pcolormesh(X, Y, Z,cmap='RdBu')
plt.title('Level sets of $f(x,y)=2x^2+3y^2-4$')
plt.show()
```

### Plotting vector fields

A vector field in the plane is an assignment of an arrow or vector $V(x,y)$ to each point of the plane. For example:
$$
V(x,y)=(\cos(x),\sin(y))
$$

Vector fields are intimately connected to differential equations, and plotting them is often useful.

```python
X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))
U = np.cos(X)
V = np.sin(Y)

plt.figure(figsize=(12,12))
plt.title('$V(x,y)=(\cos(x),\sin(y))$',fontsize=18)
Q = plt.quiver(X, Y, U, V, units='width')
qk = plt.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
plt.xlim([0,6])
plt.ylim([0,6])
```

### Simple animations

Sometimes it is useful to animate functions that depend on a parameter. The following script does this. It is better to save the code in the following cell onto a file <mark>animatetest.py</mark> and execute it after launching python from the terminal/command-line by running <mark>execfile('animatetest.py')</mark> 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def animate(t):
    line.set_ydata(np.sin(x + t/10.0))  # update the data
    return line,

# Init only required for blitting to give a clean slate.
def init():
    line.set_ydata(np.ma.array(x, mask=True))
    return line,

ani = animation.FuncAnimation(fig, animate, np.arange(1, 200), init_func=init,
                              interval=25, blit=True)
plt.show()
```
